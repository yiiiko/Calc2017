### Cauchy's Definition of the Derivative

```python
def df(a, h):
    plt.figure(figsize = (10,6), frameon=False)
    x = np.linspace(-10, 10, num=1000)
    def f(x):
        return x**2 #define the function to plot
    plt.plot(x, f(x)) #plot the parabola
    plt.plot(a, f(a), 'o', markersize = 9, color = 'black') #plot the point a, f(a)
    plt.plot(a + h, f(a+h), 'o', markersize = 9, color = 'black') #plot a point distance h from a
    slope = (f(a+h) - f(a))/(h) #compute the slope of the secant line
    plt.plot(x, slope*(x - a) + f(a)) #plot the secant line
    plt.ylim(-50, 50)
    plt.show()

interactive_plot = interactive(df, a=(-10,10,0.25), h = (0, 4, 0.1))
interactive_plot
```

```python
import numpy as np

from bokeh.layouts import row, widgetbox
from bokeh.models import CustomJS, Slider
from bokeh.plotting import figure, output_file, show, ColumnDataSource

x = np.linspace(-10, 10, 500)
y = x**2
a = np.linspace(-10,10,500)
y1 = 2*x

source = ColumnDataSource(data=dict(x=x, y=y,a=a,y1=y1))

plot = figure(x_range=(-4,4),y_range=(-10, 10), plot_width=400, plot_height=400)

plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)
plot.line('a', 'y1', source=source,  line_width=3, line_alpha=0.6)
callback = CustomJS(args=dict(source=source), code="""
    var data = source.data;
    var A = amp.value;
    var B = offset.value;
    var C = freq.value;
    x = data['x']
    y = data['y']
    a = data['a']
    y1 = data['y1']
    for (i = 0; i < x.length; i++) {
        y[i] = (x[i]**2);
         y1[i] = (2*x[i]+B);
    }
    source.change.emit();
""")

amp_slider = Slider(start=0.1, end=10, value=1, step=.1,
                    title="A", callback=callback)
callback.args["amp"] = amp_slider

freq_slider = Slider(start=-5, end=5, value=1, step=.1,
                     title="C", callback=callback)
callback.args["freq"] = freq_slider

offset_slider = Slider(start=-5, end=5, value=0, step=.1,
                       title="B", callback=callback)
callback.args["offset"] = offset_slider

layout = row(
    plot,
    widgetbox(amp_slider,offset_slider,freq_slider),
)

output_file("slider.html", title="slider.py example")

show(layout)
```

#### Lotka Volterra Predator Prey Model

$$\frac{dx}{dt} = 0.7x - 0.3xy$$

$$\frac{dy}{dt} = 0.08xy - 0.44y$$

```python
def predator_prey(a=2,b=1,c=1,d=5):
    # Initial populations
    rabbits = [10]
    foxes = [20]

    a = a/1000.
    b = b/10000.
    c = c/10000.
    d = d/1000.
    
    dt = 0.1

    N = 20000
    for i in range(N):
        R = rabbits[-1]
        F = foxes[-1]
        change_rabbits = dt * (a*R - b*R*F)
        change_foxes = dt * (c*R*F - d*F)
        rabbits.append(rabbits[-1] + change_rabbits)
        foxes.append(foxes[-1] + change_foxes)

    plt.figure(figsize=(12,6))
    plt.plot(np.arange(N+1),rabbits)
    plt.hold(True)
    plt.plot(np.arange(N+1),foxes)
    plt.ylim(0,200); plt.xlabel('Time'); plt.ylabel('Population')
    plt.legend(('Rabbits','Foxes'),loc='best');
```

```python
a = 1.
b = 0.1
c = 1.5
d = 0.75
def dX_dt(X, t=0):
    """ Return the growth rate of fox and rabbit populations. """
    return array([ a*X[0] -   b*X[0]*X[1] ,
                  -c*X[1] + d*b*X[0]*X[1] ])
```

```python
from scipy import integrate
from numpy import *
t = linspace(0, 15,  1000)              # time
X0 = array([10, 5])                     # initials conditions: 10 rabbits and 5 foxes
X = integrate.odeint(dX_dt, X0, t)
```

```python
rabbits, foxes = X.T
f1 = plt.figure(figsize=(12,6))
plt.plot(t, rabbits, 'r-', label='Rabbits')
plt.plot(t, foxes  , 'b-', label='Foxes')
plt.grid()
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('population')
plt.title('Evolution of fox and rabbit populations')
f1.savefig('rabbits_and_foxes_1.png')
```

```python
X_f0 = array([     0. ,  0.])
X_f1 = array([ c/(d*b), a/b])
all(dX_dt(X_f0) == zeros(2) ) and all(dX_dt(X_f1) == zeros(2))

values  = linspace(0.3, 0.9, 5)                          # position of X0 between X_f0 and X_f1
vcolors = plt.cm.autumn_r(linspace(0.3, 1., len(values)))  # colors for each trajectory

f2 = plt.figure(figsize=(12,6))

#-------------------------------------------------------
# plot trajectories
for v, col in zip(values, vcolors):
    X0 = v * X_f1                               # starting point
    X = integrate.odeint( dX_dt, X0, t)         # we don't need infodict here
    plt.plot( X[:,0], X[:,1], lw=3.5*v, color=col, label='X0=(%.f, %.f)' % ( X0[0], X0[1]) )

#-------------------------------------------------------
# define a grid and compute direction at each point
ymax = plt.ylim(ymin=0)[1]                        # get axis limits
xmax = plt.xlim(xmin=0)[1]
nb_points   = 20

x = linspace(0, xmax, nb_points)
y = linspace(0, ymax, nb_points)

X1 , Y1  = meshgrid(x, y)                       # create a grid
DX1, DY1 = dX_dt([X1, Y1])                      # compute growth rate on the gridt
M = (hypot(DX1, DY1))                           # Norm of the growth rate
M[ M == 0] = 1.                                 # Avoid zero division errors
DX1 /= M                                        # Normalize each arrows
DY1 /= M

#-------------------------------------------------------
# Drow direction fields, using matplotlib 's quiver function
# I choose to plot normalized arrows and to use colors to give information on
# the growth speed
plt.title('Trajectories and direction fields')
Q = plt.quiver(X1, Y1, DX1, DY1, M, pivot='mid', cmap=plt.cm.jet)
plt.xlabel('Number of rabbits')
plt.ylabel('Number of foxes')
plt.legend()
plt.grid()
plt.xlim(0, xmax)
plt.ylim(0, ymax)
f2.savefig('rabbits_and_foxes_2.png')
```
